```python
import pandas as pd

#import pandas and csv's as dataframes
neomatrix_apps_df = pd.read_csv(r"resources\raw_data\newomatrix369\extended_googleplaystore.csv")
reviews_df = pd.read_csv(r"resources\raw_data\newomatrix369\extended_googleplaystore_user_reviews.csv")
```

```python
#seprate import gauthamp cuz it HUGE
gauthamp_apps_df = pd.read_csv(r"resources\raw_data\gauthamp10\Google-Playstore.csv")
```

```python
# view initial neomatrix df
neomatrix_apps_df.head(5)
```

```python
#remove unneeded columns
clean_neo = neomatrix_apps_df[[
"App",
"Genres (categorical)",
"Rating",
"Reviews",
"Installs",
"Type",
"Price",
"Content Rating"
]]
clean_neo.rename(columns={'App': 'app', 'Genres (categorical)': 'genre', 'Rating': 'rating', \
    'Reviews': 'reviews', 'Installs': 'installs', 'Type': 'type', 'Price': 'price', \
    'Content Rating': 'content_rating'}, inplace=True)
clean_neo.head(5)
```

```python
#view review dataframe
reviews_df.head(5)
```

```python
#remove unneeded columns

clean_neo_reviews = reviews_df[[
    "App",
    "Translated_Review",
    "original_Sentiment"
]]
clean_neo_reviews_noNA = clean_neo_reviews.dropna()
clean_neo_reviews_noNA.rename(columns={'App': 'app', 'Translated_Review': 'translated_review', \
    'original_Sentiment': 'original_sentiment'}, inplace=True)
clean_neo_reviews_noNA.head(5)
```

```python
#view gauthamp dataframe
gauthamp_apps_df.head(5)
```

```python
clean_gauthamp = gauthamp_apps_df[[
    "App Name",
    "Category",
    "Rating",
    "Rating Count",
    "Installs",
    "Free",
    "Price",
    "Content Rating",
    "Ad Supported",
    "In App Purchases"
]]
clean_gauthamp.rename(columns={
    "App Name": 'app_name',
    "Category": 'category',
    "Rating": 'rating',
    "Rating Count": 'rating_count',
    "Installs": 'installs',
    "Free": 'free',
    "Price": 'price',
    "Content Rating": 'content_rating',
    "Ad Supported": 'ad_supported',
    "In App Purchases": 'in_app_purchases'
}, inplace=True)
clean_gauthamp.head(5)
```

```python
# import dependencies and postgresql login info
from sqlalchemy import create_engine
from config import login
```

```python
# connect to posgresql
engine = create_engine(f'postgresql://{login}')
```

```python
# view postgres table names
engine.table_names()
```

```python
# import clean_gauthamp into gauthamp table
clean_gauthamp.to_sql(name='gauthamp', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from gauthamp', con=engine).head()
```

```python
# import clean_neo into neomatrix table
clean_neo.to_sql(name='neomatrix', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from neomatrix', con=engine).head()
```

```python
# import clean_neo_reviews_noNA into user_reviews table
clean_neo_reviews_noNA.to_sql(name='user_reviews', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from user_reviews', con=engine).head()
```
